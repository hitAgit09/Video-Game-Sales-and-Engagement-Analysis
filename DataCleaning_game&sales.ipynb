# **Cleaning games.csv**

## *By analyzing the file i understood that it contains unnecessary column, duplicates, date stored as text, null values etc..*

```python
games = games.drop(columns=["Unnamed: 0"])
```

```python
games = games.drop_duplicates(subset = ["Title"])
```

```python
#standardizing column names
games.columns = games.columns.str.lower().str.replace(" ", "_")
```

```python
games.head()
```

## *Fix Numeric Columns Stored as Text*

```python
def convert_k(value):
    if isinstance(value, str):
        value = value.replace("K", "")
        return float(value) * 1000
    return value

for col in ["times_listed", "number_of_reviews", "plays", "playing", "backlogs", "wishlist"]:
    games[col] = games[col].apply(convert_k)
```

## *Handle Missing Values*

```python
games["rating"] = pd.to_numeric(games["rating"], errors="coerce")
games["rating"] = games["rating"].fillna(games["rating"].mean())

```

```python
#where the team is missing

games["team"] = games["team"].fillna("unknown")
```

## *Fix Date Format*

```python
games["release_date"] = pd.to_datetime(games["release_date"], errors = "coerce")
```

## *Clean genre column*

```python
games["genres"] = games["genres"].str.replace(r"[\[\]']", "", regex=True)

```

```python
games.head()
```

```python
games.to_csv("games.csv", index=False, sep=",")
```

## **Cleaning vgsales.csv**

```python
#standardizing column names

sales.columns = sales.columns.str.lower().str.replace(" ", "_")
```

```python
#Removing duplicates

sales = sales.drop_duplicates(subset=["name", "platform", "year"])
```

```python
#Handling year column

sales["year"] = pd.to_numeric(sales["year"], errors="coerce")
sales = sales.dropna(subset=["year"]).copy()
sales["year"] = sales["year"].astype(int)

```

```python
sales["publisher"] = sales["publisher"].fillna("unknown")
```

## *Checking whetther the sales columns are numeric*

```python
sales_cols = ["na_sales", "eu_sales", "jp_sales", "other_sales", "global_sales"]

for col in sales_cols:
    sales[col] = pd.to_numeric(sales[col], errors="coerce")
```

```python
sales.head()
```

```python
sales.to_csv("vgsales.csv", index=False, sep=",")
```

## **Merging Datasets**

```python
games.rename(columns={"title": "game_title"}, inplace=True)
sales.rename(columns={"name": "game_title"}, inplace=True)
```

```python
games["game_title"] = games["game_title"].str.lower().str.strip()
sales["game_title"] = sales["game_title"].str.lower().str.strip()

sales["platform"] = sales["platform"].str.lower().str.strip()
games["platform"] = games["playing"].astype(str).str.lower().str.strip()
```

```python
merged = pd.merge(
    games,
    sales,
    on="game_title",
    how="left"
)

merged.head()
```

```python
#Handling null values

for col in sales_cols:
    merged[col] = merged[col].fillna(0)
```

```python
merged.to_csv("games&sales_merged.csv", index=False, sep=",")
```
